## Variablen standardisieren

Der KNN-Klassifizierer sagt die Klasse eines gegeben Testobjekts durch seine Nähe zu anderen Beobachtungen heraus. Deshalb ist die Skala der Variablen wichtig. Variablen auf einer großen Skala (z.B. Distanz zwischen Städten in Metern) würden deshalb einen größeren Einfluss auf die Klassifizerung haben als solche mit Kleiner Skala (z.B. Tag des Monats).

Um dies zu berücksichtigen standardisieren wir nun zuerst unsere Daten:

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('TARGET CLASS', axis=1))
```

```python
scaled_features = scaler.transform(df.drop('TARGET CLASS', axis=1))
```

```python
scaled_features
```

```python
df.columns
```

```python
df_feat = pd.DataFrame(scaled_features, columns=df.columns[:-1])
```

```python
df_feat.head() # Standardisierten Werte ohne TARGET CLASS
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = scaled_features
y = df['TARGET CLASS']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```

```python
error_rate = []

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    print(pred_i != y_test) # True wenn pred_i nicht mit y_test übereinstimmt
    error_rate.append(np.mean(pred_i != y_test))  # Durchschnitt der Fehler
```

```python
np.sum(pred_i != y_test) / len(y_test) # Fehlerrate
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate, color="blue", linestyle="--", marker="o",
        markerfacecolor="red", markersize=10)
plt.title("Error Rate vs K Werte")
plt.xlabel("K")
plt.ylabel("Error Rate")
```

```python
knn = KNeighborsClassifier(n_neighbors=23)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)

print(confusion_matrix(y_test, pred))
print('\n')
print(classification_report(y_test, pred))
```
