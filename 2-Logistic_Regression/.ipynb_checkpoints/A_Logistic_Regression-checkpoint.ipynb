## Datenbereinigung

Wir wollen die fehlenden Altersdaten ersetzen anstatt die Zeilen mit fehlenden Werten einfach zu löschen. Eine Möglichkeit dies zu tun wäre das durchschnittliche Alter aller Passagiere einzufügen (Imputation). Wir können allerdings noch einen Schritt genauer vorgehen und das durchschnittliche Alter der entsprechenden Klasse einfügen. Zum Beispiel:

```python
sns.boxplot(x='Pclass', y="Age", data=train)
```

=> Preisklasse 1 sind durchschnittlich Ältere Passagiere  
=> Preisklasse 2 und 3 haben jedoch Ausreise nach oben

```python
def impute_age(cols):
    # cols = pandas Series
    print(cols)
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
sum = train[['Age', 'Pclass']].apply(np.sum,axis=1) 
```

```python
sum
```

```python
train['Age'] = train[['Age', 'Pclass']].apply(impute_age,axis=1) # Axis = 1 : Spalte
```

```python
train['Age']  # Enthält keine NaN mehr
```

```python
data = train[['Age', 'Pclass']]
s = pd.Series(list(data.T.to_numpy()), index=data.columns)
print(s)
print(type(s))
```
