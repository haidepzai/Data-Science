```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,
    n_features=2,
    centers=4,
    cluster_std=1.8,
    center_box=(-10.0, 10.0),
    shuffle=True,
    random_state=101,
    return_centers=False)
```

```python
data
```

## Daten Visualisieren

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap="rainbow")
```

```python
data[1] # Labels
```

## Cluster erzeugen

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_  # 4 Centers mit X und Y Koordinaten
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap="rainbow")
ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap="rainbow")
```

*Hinweis: Die Farbunterschiede zwischen den beiden Polts spielen hier keine Rolle. Nur die Punktzugehörigkeit zu einer Farbe.*
